```python
import pandas as pd
```

```python
bow_datar = pd.read_excel("/content/drive/MyDrive/amex/6655876ef0ec2_bowler_round1.xlsx").convert_dtypes()
```

```python
bow_datar.describe()
```

```python
bow_sort = bow_datar.sort_values(by=['ingestion_timestamp'], ascending=False)
bow_data = bow_sort.drop_duplicates(subset=['match id', 'bowler_id'])
print(bow_data.describe())
print(bow_data.info())
print(bow_data.loc[bow_data.isna()["runs"]])
print(bow_data.loc[bow_data.isna()["economy"]])
print(bow_data.loc[bow_data.isna()["wides"]])
print(bow_data.loc[bow_data.isna()["dots"]])
```

```python
bow_data.loc[bow_data['runs'].isnull(),'runs'] = round(bow_data['balls_bowled']*bow_data['economy']/6)
bow_data.loc[bow_data['economy'].isnull(),'economy'] = bow_data['runs']*6/bow_data['balls_bowled']
bow_data.loc[bow_data['wides'].isnull(), 'wides'] = round(bow_data['wides'].mean())
bow_data.loc[bow_data['noballs'] < 0, 'noballs'] = 0
bow_data.loc[bow_data['dots'].isnull(), 'dots'] = 0
bow_data['dots'] = round(bow_data['dots'].astype(float)).astype('Int64')
```
