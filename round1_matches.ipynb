```python
import pandas as pd
```

```python
mat_datar = pd.read_excel("/content/drive/MyDrive/amex/6655877834f04_matchlevel_round1.xlsx").convert_dtypes()
print(mat_datar.describe())
mat_datar.info()
mat_datar.head()
```

```python
mat_sort = mat_datar.sort_values(by=['ingestion_timestamp'], ascending=False)
mat_data = mat_sort.drop_duplicates(subset='match id')
mat_data.info()
```

```python
for index, row in mat_data[mat_data['by'].isnull()].iterrows():
    if row['winner'] == row['team1']:
        if row['toss winner'] == row['team1']:
            if row['toss decision'] == 'bat':
                mat_data.at[index, 'by'] = 'runs'
            else:
                mat_data.at[index, 'by'] = 'wickets'
        else:
            if row['toss decision'] == 'bat':
                mat_data.at[index, 'by'] = 'wickets'
            else:
                mat_data.at[index, 'by'] = 'runs'
    else:
        if row['toss winner'] == row['team2']:
            if row['toss decision'] == 'bat':
                mat_data.at[index, 'by'] = 'runs'
            else:
                mat_data.at[index, 'by'] = 'wickets'
        else:
            if row['toss decision'] == 'bat':
                mat_data.at[index, 'by'] = 'wickets'
            else:
                mat_data.at[index, 'by'] = 'runs'

```

```python
mat_data.loc[mat_data['match id'] == 8587816]['season']
```

```python
date_parts = mat_data['match_dt'].str.split("-")
mat_data['match_dt'] = (date_parts.str[1] + '/' + date_parts.str[2] + '/' + date_parts.str[0]).astype(str)
mat_data['season'] = date_parts.str[0].astype(str)
```

```python
mat_data['inning1_wickets'] = round(mat_data['inning1_wickets'].astype(float)).astype('Int64')
mat_data['inning2_wickets'] = round(mat_data['inning2_wickets'].astype(float)).astype('Int64')
mat_data.loc[mat_data['inning1_wickets'] > 10, 'inning1_wickets'] = round(mat_data.loc[mat_data['inning1_wickets'] <= 10, 'inning1_wickets'].mean())
mat_data.loc[mat_data['inning2_wickets'] > 10, 'inning2_wickets'] = round(mat_data.loc[mat_data['inning2_wickets'] <= 10, 'inning2_wickets'].mean())
mat_data.loc[mat_data['inning1_balls'] > 120, 'inning1_balls'] = round(mat_data.loc[mat_data['inning1_balls'] <= 120, 'inning1_balls'].mean())
mat_data.loc[mat_data['inning2_balls'] > 120, 'inning2_balls'] = round(mat_data.loc[mat_data['inning2_balls'] <= 120, 'inning2_balls'].mean())
mat_data.describe()
```

```python
excel_file_path = '/content/drive/My Drive/newMatches.xlsx'
mat_data.to_excel(excel_file_path, index=False)
```

```python
bow_data = pd.read_excel("/content/drive/MyDrive/newBowlers.xlsx").convert_dtypes()
bow_data.head()
```

```python
merged_df = pd.merge(bow_data, mat_data, on='match id')
filtered_df = merged_df[merged_df['wicket_count'] >= 1]
venue_counts = filtered_df.groupby('bowler_id')['ground_id'].nunique()
max_venues = venue_counts.max()
print(max_venues)
# Step 5: Find all bowlers with the maximum number of unique venues
bowlers_with_max_venues = venue_counts[venue_counts == max_venues].index.tolist()
print("Bowlers with the most unique venues:", bowlers_with_max_venues)
```
