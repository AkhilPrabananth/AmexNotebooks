```python
bat_sort = bat_datar.sort_values(by=['ingestion_timestamp'], ascending=False)
bat_data = bat_sort.drop_duplicates(subset=['match id', 'batsman_id'])
```

```python
print(bat_data.info())
bat_data.describe()
```

```python
bat_data.loc[bat_data['runs'].isnull() & bat_data['balls_faced'].isnull()]
```

```python
bat_data.loc[bat_data['runs'].isnull(), 'runs'] = round(bat_data['strike_rate']*bat_data['balls_faced']/100.0)
bat_data.loc[bat_data['strike_rate'].isnull(),'strike_rate'] = (bat_data['runs']*100.0/bat_data['balls_faced']).astype('Float64')
bat_data.loc[bat_data['Fours'].isnull(),'Fours'] = 0
bat_data.loc[bat_data['Sixes'].isnull(),'Sixes'] = 0
```

```python
bat_data.loc[bat_data['over_faced_first'] < 0, 'over_faced_first'] = bat_data.loc[bat_data['over_faced_first'] >= 0, 'over_faced_first'].mean()
```

```python
bat_data['Fours'] = round(bat_data['Fours']).astype('Int64')
bat_data['Sixes'] = round(bat_data['Sixes']).astype('Int64')
```

```python
date_parts = bat_data['batsman_dob'].str.split("-")
bat_data['batsman_dob'] = (date_parts.str[1] + '/' + date_parts.str[2] + '/' + date_parts.str[0]).astype(str)
date_parts = bat_data['bowler_dob'].str.split("-")
bat_data['bowler_dob'] = (date_parts.str[1] + '/' + date_parts.str[2] + '/' + date_parts.str[0]).astype(str)
date_parts = bat_data['match_dt'].str.split("-")
bat_data['match_dt'] = (date_parts.str[1] + '/' + date_parts.str[2] + '/' + date_parts.str[0]).astype(str)
```

```python
print(bat_data.info())
print(bat_data.describe())
bat_data.head()
```
